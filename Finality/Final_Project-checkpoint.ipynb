**1. Какие 10 лучших категорий которые устанавливаются из магазина Google Play?** \
 Берем топ 10 по количеству приложений в категориях и уже из них выявляем топ по количеству установок

```python
df["Category"].unique()
```

```python
df["Category"].value_counts().head()
```

```python
top_category = df["Category"].value_counts().reset_index().rename(columns={"Category" : "Count", "index" : "Category"})
top_category.head(10) #выявлили топ категорий по количеству приложений
```

```python
category_installs = df.groupby(["Category"])[["Installs"]].sum().reset_index()
category_installs #количество установок приложений по категориям 
```

```python
category_installs_sum_games = category_installs.replace({'Category': ('Action','Adventure','Auto & Vehicles','Arcade',\
                                        'Card','Casual','Puzzle','Racing','Role Playing','Simulation',\
                                        'Strategy','Casual','Casual','Casual')}, 'Games')\
.groupby(["Category"]).sum().sort_values('Installs', ascending= False).reset_index()
category_installs_sum_games #объеденили игровые категории в одну 
```

```python
top_category_installs = pd.merge(top_category, category_installs_sum_games, on="Category")
top_10_category_installs = category_installs_sum_games.head(10).sort_values(by=["Installs"], ascending=False)
top_10_category_installs # топ 10 категорий по установкам из топа по количеству в категории
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=60)
plt.xlabel("Category")
plt.ylabel("Number of Application")
plt.title("Top 10 Installed Category")
sns.barplot(x=top_10_category_installs["Category"], y=top_10_category_installs["Installs"])
```

**Ответ:**
1.  Games
2.  Tools
3.  Communication
4.  Productivity
5.  Photography
6.  Video Players & Editors
7.  Social
8.  Entertainment
9.  Music & Audio
10. Travel & Local

**Топ 10 категорий по количеству установок (без учета количества приложений в категории)**

```python
top_installed_app = category_installs_sum_games
top_10_installed_app = top_installed_app.sort_values(by="Installs",ascending=False).head(10)
top_10_installed_app
```

```python
plt.figure(figsize=(14,7))
plt.xticks(rotation=60)
plt.xlabel("Category")
plt.ylabel("Number of Application")
plt.title("Top 10 Installed App")
sns.barplot(x=top_10_installed_app["Category"], y=top_10_installed_app["Installs"])
```

**8. Рост количества релизов приложений с 2010 по 2020 год к году \ рост количества релизов новых андроид смартфонов** 


Приложения:

```python
df['Released'].head() #даты релиза
```

```python
released_to_datetime = pd.to_datetime(df['Released'])
```

```python
released_to_datetime
```

```python
apps_years = released_to_datetime.dropna().dt.year.value_counts(ascending = True).drop(2021)
#2021 не используем, так как данные только по июнь
apps_years #количество релизов ежегодно
```

```python
apps_years.plot(kind = 'bar') #визуализация роста количества релизов приложений ежегодно
```

```python
round(apps_years.pct_change()*100,2) #проценты увеличения количества релизов из года в год относительно предыдущего годa
```

```python
average_percentage_app = round(apps_years.pct_change().mean()*100,2)
average_percentage_app
```
